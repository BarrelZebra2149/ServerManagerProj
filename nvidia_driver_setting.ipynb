## Complete Permanent Fix Guide: RTX 3090 “GPU Stunning” / CUDA “unspecified launch failure” on Ubuntu 24.04
Your training now runs without crashing because the missing `/boot/grub` folder was blocking the kernel parameters.
Below is the full, step-by-step tutorial (copy-paste ready) so you can add any new kernel parameters in the future and make the fix 100% permanent.

```python
# Create the missing GRUB folder (one-time fix)
!sudo mkdir -p /boot/grub

# Edit GRUB to add the best kernel parameters for RTX 3090
!sudo vim /etc/default/grub

# Fix the value as
GRUB_CMDLINE_LINUX_DEFAULT="pcie_aspm=off pci=noaer pcie_port_pm=off nvidia.NVreg_EnableGpuFirmware=0"

# Enable persistence mode permanently
!sudo nvidia-smi -pm 1
```

### Apply the new parameters

```python
!sudo update-grub
!sudo reboot
```
